**Simulación de Sistema de Control de Temperatura en Almacén Industrial**

Este ejercicio simula un sistema de control para un almacén, donde se monitorean las temperaturas de tres contenedores. Si la temperatura de alguno de ellos supera un umbral crítico (40°C), el sistema activará una alerta visual después de un breve retraso simulado.




```python
# Importar las bibliotecas necesarias.
# `random` se usa para generar temperaturas aleatorias.
# `time` se usa para simular un retraso en la activación de la alerta.
import random
import time

# Definir el número de contenedores y el umbral de temperatura.
NUM_CONTENEDORES = 3
UMBRAL_ALERTA = 40 # grados Celsius

# Bucle para simular el monitoreo continuo de temperatura.
print("Iniciando monitoreo de temperaturas...")

# Ejecutar este ciclo varias veces para ver diferentes escenarios.
for ciclo in range(1, 4): # Simular 3 ciclos de monitoreo para demostración
    print(f"\n--- Ciclo de Monitoreo {ciclo} ---")

    temperaturas = []
    # Generar una temperatura aleatoria para cada contenedor.
    for i in range(NUM_CONTENEDORES):
        # Las temperaturas se generan entre 25 y 45 grados Celsius para simular variaciones.
        temp = random.uniform(25.0, 45.0)
        temperaturas.append(temp)
        print(f"Contenedor {i+1}: {temp:.2f}°C")

    # Verificar si alguna temperatura supera el umbral de alerta.
    alerta_activa = False
    for i, temp in enumerate(temperaturas):
        if temp > UMBRAL_ALERTA:
            print(f"¡ALERTA! La temperatura del Contenedor {i+1} ({temp:.2f}°C) ha superado los {UMBRAL_ALERTA}°C.")
            alerta_activa = True

    # Si se activó una alerta, simulamos un retraso antes de la "alerta visual".
    if alerta_activa:
        print("Activando alerta visual... (simulando retraso de 2 segundos)")
        time.sleep(2) # Pausa la ejecución por 2 segundos para simular el retraso
        print("¡Alerta visual activada!")
    else:
        print("Todas las temperaturas están dentro del rango normal.")

print("\nMonitoreo de temperaturas finalizado.")
```

```python
# Importar la biblioteca `statistics` para calcular el promedio fácilmente.
import statistics

# Definir las especificaciones del perno.
DIAMETRO_IDEAL = 10.0 # Diámetro objetivo en mm
TOLERANCIA = 0.2    # Tolerancia permitida en mm (+/- 0.2 mm)
NUM_PERNOS = 4      # Número de pernos a medir

# Calcular los límites de aceptación.
LIMITE_INFERIOR = DIAMETRO_IDEAL - TOLERANCIA
LIMITE_SUPERIOR = DIAMETRO_IDEAL + TOLERANCIA

print(f"--- Control de Calidad de Pernos ---")
print(f"Diámetro ideal: {DIAMETRO_IDEAL} mm")
print(f"Tolerancia: +/- {TOLERANCIA} mm")
print(f"Rango aceptable: {LIMITE_INFERIOR} mm a {LIMITE_SUPERIOR} mm\n")

# Listas para almacenar las medidas y los pernos aceptados/rechazados.
medidas_pernos = []
pernos_aceptados = []
pernos_rechazados = []

# Ciclo para recibir las medidas de cada perno.
for i in range(NUM_PERNOS):
    while True:
        try:
            medida = float(input(f"Ingrese la medida del perno {i+1} (en mm): "))
            medidas_pernos.append(medida)
            break # Salimos del bucle si la entrada es válida
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número.")

# Procesar cada perno para determinar si es aceptado o rechazado.
print("\n--- Resultados del Control de Calidad ---")
for i, medida in enumerate(medidas_pernos):
    if LIMITE_INFERIOR <= medida <= LIMITE_SUPERIOR:
        print(f"Perno {i+1} ({medida:.2f} mm): ACEPTADO")
        pernos_aceptados.append(medida)
    else:
        print(f"Perno {i+1} ({medida:.2f} mm): RECHAZADO (fuera de rango)")
        pernos_rechazados.append(medida)

# Calcular el promedio de los pernos aceptados.
if pernos_aceptados:
    promedio_aceptados = statistics.mean(pernos_aceptados)
    print(f"\nPromedio de los pernos ACEPTADOS: {promedio_aceptados:.2f} mm")
else:
    print("\nNo se aceptaron pernos, no se puede calcular un promedio.")

print(f"Total de pernos aceptados: {len(pernos_aceptados)}")
print(f"Total de pernos rechazados: {len(pernos_rechazados)}")
print("\nControl de calidad finalizado.")
```
